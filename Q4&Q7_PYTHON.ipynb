```python
import pandas as pd
```

```python
### Reading the 'json' file with read_json() and storing it in 'data' variable
data=pd.read_json('https://data.nasa.gov/resource/y77d-th95.json')
```

```python
data.head(3)
```

```python
### Dropping columns ':@computed_region_cbhk_fwbd' and	':@computed_region_nnqa_25f4' as they are not essential 
### as per our problem statement
data = data.drop(labels=[':@computed_region_cbhk_fwbd',':@computed_region_nnqa_25f4'],axis=1)
```

```python
data['year'].isnull().value_counts()
```

```python
###In the pandas library the minimum representable datetime value is jan 1,1583.Hence the years preceding '1583' are filled with 0 and the years of 
### all the remaining records are extracted
data['year']=pd.to_datetime(data['year'],errors='coerce').dt.year.fillna('0').astype('int')
```

```python
data['geolocation'].isnull().value_counts()
```

```python
### As per the problem statement let us extract the point coordinates from the 'geolocation columns and store it in separate column 'point_coordinates''
data['point_coordinates']=data['geolocation'].str.get('coordinates')
```

```python
data.head()
```

```python
### As per the problem statement , the 'geolocation' column is no longer required, hence dropping it
data = data.drop(labels=['geolocation'],axis=1)
```

```python
data.head(3)
```

```python
data.info()
```

```python
### Converting the dataframe to the csv file format
data.to_csv('converted_to_csv.csv',index=False)
```
